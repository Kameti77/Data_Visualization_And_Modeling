## 1. Dataset Summary

This dataset was downloaded from Kaggle. It contains information about the global AI and machine learning job market. The dataset includes more than 15,000 job listings from over 50 countries.

The data was created for educational and research purposes. It was generated based on industry research and real job market trends to simulate how AI jobs, salaries, and hiring patterns look around the world.

The dataset includes several features such as job title, salary in USD, experience level, employment type, company location, company size, employee residence, and remote work ratio. These features help show how different factors may affect salaries and job opportunities in the AI field.

This dataset is interesting to me because I would like to explore machine learning in the future.
```python
import pandas as pd

df = pd.read_csv("ai_job_dataset.csv")

df.head(10
       )
```

## Question 1
### What is the average salary of AI jobs in this dataset?

```python
average_salary = df['salary_usd'].mean()
average_salary
```

**Answer**

The average salary of AI jobs in this dataset is approximately **$115,349 per year**. This shows that careers in artificial intelligence and machine learning tend to offer relatively high salaries compared to many other industries.

## Question 2
### Which job titles appear most frequently in the dataset?

```python
df['job_title'].value_counts().head(10)
```

**Answer**

The most common job title in the dataset is **Machine Learning Researcher**, with **808 job postings**. Other common positions include **AI Software Engineer (784)**, **Autonomous Systems Engineer (777)**, **Machine Learning Engineer (772)**, and **AI Architect (771)**. This suggests that companies are actively hiring engineers and researchers to develop AI technologies.


## Question 3
### How does salary differ by experience level?

```python
df.groupby('experience_level')['salary_usd'].mean()
```

**Answer**

The analysis shows that salary increases as experience level increases.

- **Entry-level (EN):** about **$63,133**
- **Mid-level (MI):** about **$87,955**
- **Senior-level (SE):** about **$122,188**
- **Executive-level (EX):** about **$187,724**

This indicates that experience plays a major role in determining salary in the AI job market.


## Question 4
### Which countries have the highest number of AI job postings?

```python
df['company_location'].value_counts().head(10)
```

**Answer**

The country with the highest number of AI job postings is **Germany with 814 listings**. Other countries with many AI job opportunities include **Denmark (778)**, **France (769)**, **Canada (769)**, and **Austria (765)**. This suggests that AI job opportunities are distributed across several countries with strong technology sectors.


## Question 5
### Do remote jobs pay more or less than non-remote jobs?

```python
df.groupby('remote_ratio')['salary_usd'].mean()
```

**Answer**

The results show that salaries are relatively similar across different remote work levels.

- **On-site jobs (0% remote):** about **$114,140**
- **Hybrid jobs (50% remote):** about **$115,777**
- **Fully remote jobs (100% remote):** about **$116,161**

This suggests that remote work does not significantly reduce salaries in AI jobs, and fully remote jobs may even offer slightly higher salaries on average.
