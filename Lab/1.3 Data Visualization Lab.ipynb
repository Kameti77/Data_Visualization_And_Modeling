## Design of a âGuaranteed Smash Hitâ

Based on the distributions observed in the dataset, a song with the most common and typical values of these features would have the following characteristics:

- **BPM:** 110â130 BPM  
- **Key:** C#  
- **Mode:** Major  
- **Danceability:** 70â80%  
- **Energy:** 60â75%  
- **Speechiness:** 2â10%  
- **Acousticness:** 0â25%

These values represent the most common ranges found among songs in the dataset. A song that falls within these ranges would match the typical characteristics of popular songs and therefore has a higher chance of becoming a smash hit according to the data.


## Find the actual smash hits

```python
hits = df[
    (df['bpm'] >= 110) & (df['bpm'] <= 130) &
    (df['key'] == 'C#') &
    (df['mode'] == 'Major') &
    (df['danceability_%'] >= 70) & (df['danceability_%'] <= 80) &
    (df['energy_%'] >= 60) & (df['energy_%'] <= 75) &
    (df['speechiness_%'] >= 2) & (df['speechiness_%'] <= 10) &
    (df['acousticness_%'] >= 0) & (df['acousticness_%'] <= 25)
]

hits[['track_name','artist(s)_name','bpm','danceability_%','energy_%']]
```
